## Deep Neural Network Representaion
- NN contains of input layer, hidden layers, and output layer
-  We use $L$ to denote the number of layers in a NN.
- $n[l]$ denotes the number of neurons in a specific layer $l$
- Usually, the input layer is $0th$ layer, while the output layer is $Lth$ layer, if the number of layers is $L$. 


## Image repsentation:
- Each image is represented by a numpay-array with shape(height, width, 3). For convenience, the image is  reshaped to a numpy-array of shape (height*width *3, 1)
- By vetorizing, the input matrix has a shape(height*width*3, m ) , m is the number of examples

## Getting your matrix dimension right

   - Dimension of $W$ is (n[l],n[l-1])
   - Dimension of $b$ is (n[l],1)
   - Dimension of $Z[L],A[L],dZ[l]$ and $dA[l]$ is (n[l],m)

## Building blocks of deep neural networks
- Forward propagation for a layer $l$:
    - Input $A[l-1]$
    -  Linear Forward: $z[l] = W[l]A[L-1]+b[l]$
    - Activation: $A[l] = g(z[l])$
    - Cache $Z[l],A[l-1]$, output $A[l]$
- Back propagation for layer $l$ 
    - Input $dA[l]$, caches
    - $dz[l]= dA[l] * g'(z[l])$
    - $dw[l] = dz[l] * A[l-1].T$/m
    - $db[l] = sum(dz[l])/m$  - Don't forget axis =1 , keepdims = True
    - $dA[l-1] = w[l].T * dZ[l]$ - The multiplication are a dot product
    - Output $dA[l-1],dw[l],db[l]$

## Parameters and Hyperparameters
- Parameters are $w$ and $b$
- Hyper parameters are :
    - Learning rate
    - Number of iteration
    - Number of hidden layers 
    - Number of hidden units
    - Choice of activaiton functions
   
## General methodology to build a NN:
- Define the neural network strcture
- Initialize the model's parameters
- Loop:
    - Implement forward propagation
    - Compute loss
    - Implement backward propagation to get the gradients
    - Update parameters(gradient descent)
- Use trained parameter to predict lables
 
## Implementation
- For linear forward, input $(A[l-1], W[l],b[l])$, return $Z[l]$, cache $A[l-1], W[l], b[l]$
- For activation forward, input $Z[l]$, return $A[l]$, cache$Z[l]$
